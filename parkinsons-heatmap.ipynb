```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
import seaborn as sb

# Dictates size of graph.
pt.figure(figsize=(16, 6))

# Converts file to dataframe.
df_raw = pd.read_csv("parkinsons.data")

# Removes the status column and inserts it in the front of the graph, for easy accessibility when analyzing.
first_column = df_raw.pop("status")
df_raw.insert(0, "status", first_column)

# Creates heatmap using correlation data and turning on data annotation.
sb.heatmap(df_raw.corr(), annot=True)
```

```python
# Dictates size of graph.
pt.figure(figsize=(16, 6))

# Creates new dataframe that filters data that has a status of 1.
df_edit1 = df_raw[df_raw["status"] == 1]

# Creates heatmap of new dataframe using correlation data and turning on data annotation.
sb.heatmap(df_edit1.corr(), annot=True)
```

```python
# Dictates size of graph.
pt.figure(figsize=(16, 6))

# Creates new dataframe that filters data that has a status of 0.
df_edit0 = df_raw[df_raw["status"] == 0]

# Creates heatmap of new dataframe using correlation data and turning on data annotation.
sb.heatmap(df_edit0.corr(), annot=True)
```
