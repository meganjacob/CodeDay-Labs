```python
import pandas as pd
import sklearn as sk
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
# Establishes dataframe without name column.
df_raw = pd.read_csv("parkinsons.data")
df_raw.pop("name")

# Defines input and output data.
y = df_raw.iloc[:,16]
X = df_raw.iloc[:,:16]

# Creates testing set from the first 100 data points.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.51, random_state = 42)

# Trains and predicts through a random forest classifier.
reg_rf = RandomForestClassifier()
reg_rf.fit(X_train, y_train)
y_pred = reg_rf.predict(X_test)

# Prints accuracy report.
print(metrics.classification_report(y_test, y_pred))

# Prints testing set dataframe with status and prediction.
df_raw = df_raw.head(100)
df_raw["prediction"] = y_pred
first_column = df_raw.pop("status")
df_raw.insert(0, "status", first_column)
second_column = df_raw.pop("prediction")
df_raw.insert(1, "prediction", second_column)
print(df_raw)
```
